```python
import pandas as pd
import numpy as np
import seaborn as sns

sns.set(rc={'figure.figsize':(14,8)})
```

```python
au_df = pd.read_csv('sales_country.csv', header='infer')
au_df['Month'] = pd.to_datetime(au_df['Month'])
display(au_df.head())

ax = sns.lineplot(x='Month', y='Active Users', hue='Country', data=au_df)
ax.set(ylabel='Active Users in Thousands', ylim=[0,1500])
```
