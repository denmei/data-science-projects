```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.ticker as mtick
import matplotlib.ticker as ticker


sns.set_style('darkgrid')
sns.set_style({'font.family': 'Montserrat'})
sns.set_context("notebook")
```

```python
# n-Day Retention

days = []
for x in range(0, 15):
    days += [x]
    
cohort_1 = [100.0, 14.0, 12.0, 10.5, 8.0, 7.5, 7.3, 7.1, 6.8, 6.5, 6.5, 6.2, 5.0, 4.6, 4.5]
cohort_2 = [100, 22, 19, 18, 16, 14.50, 12, 10.50, 9.50, 9, 8.50, 8.10, 7.80, 7.50, 7.4]
n_day1_df = pd.DataFrame()
n_day1_df['retention'] = cohort_1
n_day1_df['Cohort'] = 'Cohort 1'
n_day1_df['days'] = days

n_day2_df = pd.DataFrame()
n_day2_df['retention'] = cohort_2
n_day2_df['Cohort'] = 'Cohort 2'
n_day2_df['days'] = days

n_day_df = n_day1_df.append(n_day2_df)
fig, ax = plt.subplots(figsize=[12,6])
sns.lineplot(x='days', y='retention', hue='Cohort', data=n_day_df, ax=ax, marker="o", linewidth=2.5)
ax.set(ylabel='% Users coming back', xlabel='Days since App Download', title='N-Day Retention')
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))

plt.tight_layout()
```

```python
days = ['Day 0', 'Days 1-3', 'Days 4-6', 'Days 7-14', 'Days 15-30']
cohort_1 = [100.0, 14.0, 12.0, 10.5, 9]
cohort_2 = [100, 22, 19, 18, 16]
n_day1_df = pd.DataFrame()
n_day1_df['retention'] = cohort_1
n_day1_df['Cohort'] = 'Cohort 1'
n_day1_df['days'] = days
n_day1_df = n_day1_df.reset_index()

n_day2_df = pd.DataFrame()
n_day2_df['retention'] = cohort_2
n_day2_df['Cohort'] = 'Cohort 2'
n_day2_df['days'] = days
n_day2_df = n_day2_df.reset_index()

n_day_df = n_day1_df.append(n_day2_df)
n_day_df['i'] = n_day_df.index
n_day_df = n_day_df.sort_values('i')
fig, ax = plt.subplots(figsize=[12,6])
sns.lineplot(x='days', y='retention', hue='Cohort', data=n_day_df, ax=ax, marker="o", linewidth=2.5, sort=False)
ax.set(ylabel='% Users coming back', xlabel='Days since App Download', title='Bracket Retention')
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.xaxis.set_major_locator(ticker.MultipleLocator(1))

plt.tight_layout()
```

```python
# churn rate
months = ['Dec-2018', 'Jan-2019', 'Feb-2019', 'Mar-2019', 'Apr-2019', 'May-2019', 'Jun-2019', 'Jul-2019', \
         'Aug-2019', 'Sep-2019']
    
cohort_1 = [0.3, 0.33, 0.34, 0.31, 0.32, 0.33, 0.26, 0.27, 0.25, 0.23]
n_day1_df = pd.DataFrame()
n_day1_df['churn'] = cohort_1 
n_day1_df['churn'] = n_day1_df['churn'] * 100
n_day1_df['Cohort'] = 'Cohort 1'
n_day1_df['months'] = months

fig, ax = plt.subplots(figsize=[12,6])
sns.lineplot(x='months', y='churn', data=n_day1_df, ax=ax, sort=False)
ax.set(ylim=[0,50], xlabel='Month', ylabel='% active customers churned')
ax.axvline(5, ls='--', color='C1')
ax.yaxis.set_major_formatter(mtick.PercentFormatter())
ax.text(5.1, 40, "Release of new Feature", horizontalalignment='left', size='medium', color='C1')
plt.tight_layout()
```
